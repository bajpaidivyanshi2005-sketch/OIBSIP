```python
import pandas as pd

df = pd.read_csv("customer_shopping_data (1).csv")
df.head()
```

# **BASIC OVERVIEW**

```python
print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nInfo:\n")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

# **SUMMARY STATISTICS**

```python
print("\nSummary:\n", df.describe(include="all"))
```

# **ADD A TOTAL SPEND COLUMN**

```python
df["total_spend"] = df["quantity"] * df["price"]
```

# **CONVERT INVOICE_DATE TO DATETIME**

```python
df["invoice_date"] = pd.to_datetime(df["invoice_date"], dayfirst=True)
```

# **EXPLORATORY ANALYSIS**

Gender Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="gender", palette="Set2")
plt.title("Gender Distribution")
plt.show()
```

Age Distribution

```python
sns.histplot(data=df, x="age", bins=30, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.show()
```

Payment Method Distribution

```python
sns.countplot(data=df, x="payment_method", palette="Set3")
plt.title("Payment Method Usage")
plt.show()
```

Product Category Popularity

```python
sns.countplot(data=df, y="category", order=df["category"].value_counts().index, palette="coolwarm")
plt.title("Top Purchased Categories")
plt.show()
```

Shopping Mall Activity

```python
top_malls = df["shopping_mall"].value_counts().head(10)
sns.barplot(x=top_malls.values, y=top_malls.index, palette="viridis")
plt.title("Top 10 Shopping Malls by Transactions")
plt.xlabel("Number of Transactions")
plt.show()
```

Age Vs. Spending

```python
sns.scatterplot(data=df, x="age", y="total_spend", hue="gender", alpha=0.5)
plt.title("Age vs Total Spending")
plt.show()
```

Time-Based Analysis

```python
df["month"] = df["invoice_date"].dt.to_period("M")
monthly_sales = df.groupby("month")["total_spend"].sum().reset_index()
monthly_sales["month"] = monthly_sales["month"].astype(str)

sns.lineplot(data=monthly_sales, x="month", y="total_spend", marker="o")
plt.title("Monthly Total Sales")
plt.xticks(rotation=45)
plt.show()
```

Average Spend By Gender

```python
avg_spend = df.groupby("gender")["total_spend"].mean().reset_index()
sns.barplot(data=avg_spend, x="gender", y="total_spend", palette="Set2")
plt.title("Average Spend by Gender")
plt.ylabel("Average Total Spend")
plt.show()
```
